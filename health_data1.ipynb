```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('dirty_v3_path.csv')
```

```python
df.columns=df.columns.str.lower()
```

```python
df.columns=df.columns.str.replace(" ","_")
```

```python
df.columns
```

```python
df=df.drop(['random_notes','noise_col'],axis=1)
```

```python
df['medical_condition'].fillna("Unknown", inplace=True)
df.head()
```

```python
df['age'].fillna(df['age'].median(),inplace=True)
df.head()
```

```python
df['gender'].fillna(df['gender'].mode()[0],inplace=True)
df.head(10)
```

```python
df['glucose'].fillna(df['glucose'].median(),inplace=True)
df['blood_pressure'].fillna(df['blood_pressure'].median(),inplace=True)
df.head(20)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df=df[(df['bmi']>10) &(df['bmi']<60)]
df=df[(df['oxygen_saturation']<=100)]
df=df[(df['sleep_hours']<24)]
df.head()
```

```python
df.info()
```

```python
# BMI_category

df['bmi_category']=pd.cut(df['bmi'],bins=[0,18.5,25,30,60],
labels=['Underweight','Normal','Overweight','obese'])
```

```python
df['risk_score']=(df['glucose']*0.25+
                  df['bmi']*0.20+
                  df['cholesterol']*0.20+
                  df['stress_level']*0.15+
                  df['hba1c']*0.20)
```

```python
df.head()
```

```python
# Check unrealistic medical values

def validation_report(df):
    report={}
    report['nagetive_glucose']=(df['glucose']<0).sum()
    report['bmi>60']=(df['bmi']>60).sum()
    report['oxygen>100']=(df['oxygen_saturation']>100).sum()
    report['sleep>24']=(df['sleep_hours']>24).sum()
    return report
validation_report(df)
```

```python
import seaborn as sns
import pandas as pd
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True),annot=False,cmap='coolwarm')
plt.show()
```

```python
# does smoking increase cholestrol
df.groupby('smoking')['cholesterol'].mean()
```

```python
from scipy.stats import ttest_ind
smoker=df[df['smoking']==1]['cholesterol']
non_smokers=df[df['smoking']==0]['cholesterol']
ttest_ind(smoker,non_smokers)
```

```python
# does stress reduce sleep
df.groupby(pd.cut(df['stress_level'],bins=4))['sleep_hours'].mean()
```

```python
# compare condition
df.groupby('medical_condition')['risk_score'].mean().sort_values(ascending=False)
```

```python
# risk sagmentation
df['risk_level']=pd.qcut(df['risk_score'],3,labels=['low','medium','high'])
df.groupby('risk_level')['lengthofstay'].mean()
df.head()
```
